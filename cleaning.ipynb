```python
import pandas as pd
```

```python
df = pd.read_csv('csv_results/judges_fullname_circuits.csv')
df.head()
```

```python
df['circuit_join'] = df['circuit_join'].str.upper()
df.to_csv('csv_results/judges_fullname_circuits.csv', index=False)
```

```python
df['president'].value_counts()
```

```python
D = ['William J. Clinton', 'Barack Obama', 'Joseph R. Biden', 'Jimmy Carter']
R = ["Donald J. Trump", 'George W. Bush', 'Ronald Reagan', 'George H.W. Bush', 'Richard M. Nixon', 'Gerald Ford']
df['party'] = df['president'].apply(lambda x: 'D' if x in D else 'R')
```

```python
df_party = pd.read_csv('csv_results/party.csv')
```

```python
df_party.dtypes
```

```python
new_df = pd.read_csv('csv_results/analysis.csv')
```

```python
new_df = df_party.merge(df_president, on='circuit_join')
```

```python
# rename columns "Barack Obama" to "Obama", "William J. Clinton" to "Clinton",  "Joseph R. Biden" to "Biden", "Donald J. Trump" to "Trump", "George W. Bush" to "Bush", "Ronald Reagan" to "Reagan", "George H.W. Bush" to "HWBush", "Richard M. Nixon" to "Nixon", "Gerald Ford" to "Ford"
new_df.columns = new_df.columns.str.replace('Barack Obama', 'Obama')
new_df.columns = new_df.columns.str.replace('William J. Clinton', 'Clinton')
new_df.columns = new_df.columns.str.replace('Joseph R. Biden', 'Biden')
new_df.columns = new_df.columns.str.replace('Donald J. Trump', 'Trump')
new_df.columns = new_df.columns.str.replace('George W. Bush', 'Bush')
new_df.columns = new_df.columns.str.replace('Ronald Reagan', 'Reagan')
new_df.columns = new_df.columns.str.replace('George H.W. Bush', 'HWBush')
new_df.columns = new_df.columns.str.replace('Richard M. Nixon', 'Nixon')
new_df.columns = new_df.columns.str.replace('Gerald Ford', 'Ford')
new_df.columns = new_df.columns.str.replace('Jimmy Carter', 'Carter')

new_df
```

```python
new_df.head()
```

```python
# if D > R, then D, else if R > D, then R, else if D == R, then NA
new_df['party'] = new_df.apply(lambda x: 'D' if x.D > x.R else 'R' if x.R > x.D else 'NA', axis=1)
```

```python
new_df.to_csv('csv_results/analysis.csv', index=False)
```
